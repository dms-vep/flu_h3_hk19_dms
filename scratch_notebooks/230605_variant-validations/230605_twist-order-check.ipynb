```python
import pandas as pd
import Bio
from Bio.Seq import Seq
```

```python
variants = pd.read_csv('230602_twist_hk19-validation-order.csv')
variants.columns = ['seq_name', 'seq']
variants
```

```python
def check_aa_residue(full_seq, site):
    ecto_seq = Seq(full_seq[20:1676])
    ecto_aa = ecto_seq.translate()
    
    return ecto_aa[site-1]
```

```python
target_site = [50, 156, 189, 124, 193, 193, 131, 160, 157, 159, 159]
aa_list = []
i = 0

for sequence in variants['seq']:
    aa = check_aa_residue(sequence, target_site[i])
    aa_list.append(aa)   
    i+=1                 
```

```python
variants['target_aa_id'] = aa_list
variants
```
