```python
data_run2_7 = (
    pd.read_excel('../experiments/neut_assays/neut_data/2022-01-19_HK19_redone_H6R235M_AG_AUSAB_tidy.xlsx')
      .query('serum == "07-AUSAB"')
      .query('virus == "HK19_WT"')
)

data_run1_7 = (
    pd.read_excel('../experiments/neut_assays/neut_data/2022-01-14-HK19_AlexGreninger_AUSAB01-16_tidy.xlsx')
      .query('serum == "07-AUSAB"')
      .query('virus == "HK19_WT"')
      .query('concentration != 0.000390625') # drop this, measured in second run already
)

data_7 = data_run1_7.append(data_run2_7)
data_7.head()
```

```python
fits_7 = neutcurve.CurveFits(data_7)
fits_7.fitParams(ics=[65,99])
```

```python
fig, axes = fits_7.plotSera(
                xlabel='serum dilution',
                virus_to_color_marker={
                    'HK19_WT': ('#999999', 'o')}, 
#                 vlines={'07-AUSAB': [{'x': 0.000158, 'color': '#FB9A29'}, {'x': 0.000395, 'color':'#EC7014'}, 
#                                      {'x': 0.00158, 'color': '#CC4C02'}, {'x': 0.00632, 'color':'#993404'}, 
#                                      {'x': 0.0158, 'color':'#662506'}],
#                         },
                vlines={'07-AUSAB': [{'x': 0.000158, 'color': '#FEC44F'}, {'x': 0.000395, 'color':'#FB9A29'}, 
                                     {'x': 0.00158, 'color': '#EC7014'}, {'x': 0.00632, 'color':'#CC4C02'}, 
                                     {'x': 0.0158, 'color':'#993404'}],
                        },
                )

bbox = matplotlib.transforms.Bbox([[0, 0], [5.5, 3.3]])

plt.savefig('scratch_figs/ser07_neut_curve_ic.pdf', bbox_inches = bbox)
```

```python
dilutions = [0.000158, 0.000395, 0.00158, 0.00632, 0.0158]

for dilution in dilutions:
    ic = 100 * (1 - (fits_7.getCurve(serum='07-AUSAB', virus='HK19_WT', replicate='average').fracinfectivity(dilution)))
    ic = str(ic)[0:5]
    print(f'{dilution}: IC{ic}')
```
