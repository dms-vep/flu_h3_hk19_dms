## Analyzing mutation frequency across variants

Make plots where the y-axis is **maximum** and **mean** frequency of specific **mutations** at a site. Additionally, check how many of these mutations are off-target. Break down analysis across epitopes versus general mutations.

Start by getting the dataframe of mutation counts, and adding a column defining whether that site is a major epitope or not.

```python
# Load epitopes
epitopes = [212, 161, 154, 157, 179, 205, 178, 280, 140, 147, 150, 163, 217, 190, 209, 216, 81, 211, 102, 111, 330, 345]
paired_epitopes = [150, 154, 157, 161, 163, 178, 179, 209, 211, 212, 216, 217]

print('all epitopes:',epitopes,'\n','\n','paired epitopes:',paired_epitopes)
```

Note that there are 22 total epitope sites, and 12 of these are included in both the paired primers and single primers, while 10 are only present as singles. 

```python
# get df with a column defining epitope sites
mut_counts = variants.mutCounts(variant_type='all', mut_type='aa', samples=None, 
                                libraries=variants.libraries).rename(columns={'count': 'mut_count'})

def assign_epitope(x, paired_epitopes, epitopes):
    if x in paired_epitopes:
        return 'paired_epitope'
    elif x in epitopes:
        return 'epitope'
    else: return 'other'

mut_counts['epitope'] = mut_counts.apply(lambda x: assign_epitope(x['site'], paired_epitopes, epitopes), axis=1)

# order to make data read more cleanly
mut_counts['epitope'] = pd.Categorical(mut_counts['epitope'], categories=['other','epitope','paired_epitope'], 
                                       ordered=True)
```

To clarify here, if a site exists in the paired primers, it is called `paired_epitope`. It **also** still exists in the single epitope primers. But the paired definition takes precedence.

Then, add 'off-target' labels to the dataframe. Targeted mutations are called based on primer names in the targeted_primers dataset, generated in the dir 'library_design' [here]('library_design/targeted_library_primers/results/hk19_primers.csv'). Note that these primers are numbered sequentially from the first mutated site. However, 'site 1' corresponds to site 22 in chimeric sequential numbering. So we need to add **21** to each site number so it matches up with the chimeric numbering.

```python
# load targeted primer dataset
targeted_primers = pd.read_csv('scratch_notebooks/hk19_primers.csv',
                              header=None).rename(columns={0:'name', 1:'primer'})
```

```python
def get_targeted_sites(primer_name, re_pattern, targeted_muts_list):
    primer_name = re_pattern.sub('', primer_name)
    site_number = int((re.findall(r'\d+', primer_name))[0])
    chimeric_site = site_number + 21
    aa_mut = primer_name[-1]
    
    targeted_muts_list.append(str(chimeric_site)+aa_mut)
    return targeted_muts_list
```

```python
# extract a list of targeted sites based on primer names
re_pattern = re.compile(r'^(.*mut)')
targeted_muts = []
targeted_primers.apply(lambda x: get_targeted_sites(
    x['name'], re_pattern, targeted_muts), axis=1)

targeted_muts[:10]
```

```python
# add column that matches 'site#_AA' format of list
mut_counts['stripped_mut'] = mut_counts['mutation'].str.replace(r'^[A-Z]', '')

# add 'targeted' and 'off-target' labels
# anything labeled 'paired_epitope' or 'epitope' is by definition targeted by random epitope primers
mut_counts['targeted'] = np.where(
    (~mut_counts['stripped_mut'].isin(targeted_muts)) & (mut_counts['epitope'] == 'other'), 'off-target', 
    'targeted')

# mut_counts = mut_counts.drop(['stripped_mut'], axis=1)

mut_counts
```

Now we have a dataframe that includes columns defining epitope identity of that site, and whether it was targeted in our set of mutagenic primers. We can now move on to analyzing mutation distributions across library variants.

### How many backgrounds is each mutant represented in, across the library?

This is important because, in downstream epistasis modeling, we will likely get more accurate results if we have data on how mutation effects change across **lots** of different variant backgrounds.

For now, just plot the backgrounds for LibB, as this library has more sequenced variants and we've decided to use LibB in downstream experiments.

```python
targeted_muts = mut_counts.loc[(mut_counts['targeted'] == 'targeted')]

# libB_missed_muts = libB_mut_counts[~libB_mut_counts['stripped_mut'].isin(targeted_muts)]

# libB_h6_dup_bc = libB_df[libB_df['barcode'].isin(h6_barcodes)]
```

```python
p = (
    ggplot(targeted_muts,
           aes(x='mut_count')) + 
    geom_histogram() + 
    theme(figure_size=(12, 5),
         panel_grid_major_x=element_blank(),
         ) + 
    labs(x='unique_background_count', 
         y='number_muts', title=f'variant background spread') +
    scale_fill_manual(values=CBPALETTE[2:]))

_ = p.draw()

plotfile = 'scratch_notebooks/scratch_figs/var_bkgd_spread.pdf'

# os.path.join(config['figs_dir'],
#                         f"{lib_id}_var_bkgd_spread.pdf")
print(f"Saving plot to {plotfile}")
p.save(plotfile)
```

This skew looks really good. On average, most mutations are represented in ~50 variant backgrounds. There are a handful represented more sparsely, and then a long tail of mutations represented in **lots** of backgrounds. These are probably the targeted epitopes.

### How many off-target mutations are present?

```python
mut_counts_agg = mut_counts.groupby(['library', 'targeted']).aggregate({'mut_count': 'sum'})
mut_counts_agg.groupby(level=0).apply(lambda x: 100 * x / x.sum()).rename(columns={'mut_count': 'fraction_muts'})
```

This looks pretty good. For each library, we only have ~5% off-target mutations. Note that 'targeted' encompasses all of the random epitope mutations, though, which were at very high levels and include **every** possible amino acid change. So this could be skewing our analysis of 'off-target' levels. If we look more closely at the non-epitope mutations:

```python
targeted_muts_df = mut_counts[mut_counts.epitope.isin(['other'])]

targeted_muts_agg = targeted_muts_df.groupby(['library', 'targeted']).aggregate({'mut_count': 'sum'})
targeted_muts_agg.groupby(level=0).apply(lambda x: 100 * x / x.sum()).rename(columns={'mut_count': 'fraction_muts'})
```

Still looks good! Even if we ignore the random epitope mutations, we still only have ~7% off-target mutations compared to mutations generated by targeted primers.

### Analyze mutation frequency

Next, we look more closely at stats for specific mutations at each site, rather than just cumulative sum. Based on number of total variants per library, we can get the **frequency** of mutation at each site. Then we will also look at the maximum frequency of a single mutation seen at each site. I originally looked at the mean frequency per site as well, but it's difficult to normalize this because there's a different number of targeted mutations at each site.

```python
# define the total variants in each library
total_variants = variants.n_variants_df(samples=None, libraries=variants.libraries)
total_variants = total_variants.rename(columns={'count': 'variant_count'})
```

```python
# set up a df with separate columns for the stats of interest
mut_stats_targeted = (mut_counts
             .groupby(['library', 'epitope', 'site', 'targeted'])
             .aggregate(mut_count_total=('mut_count', 'sum'), 
                        max_mut_count=('mut_count', 'max')) 
             .reset_index()
             .merge(total_variants, how='left', on='library')
             .drop(['sample'], axis=1)
             .assign(mut_freq_total=lambda x: x.mut_count_total / x.variant_count)
             .assign(max_mut_freq=lambda x: x.max_mut_count / x.variant_count))

mut_stats_targeted.head()
```

```python
for stat in ('mut_freq_total', 'max_mut_freq'):
    p = (
        ggplot(
            mut_stats_targeted,
            aes(x='site', y=stat, fill='epitope')) + 
        geom_bar(stat='identity', width=5) + 
        theme(figure_size=(10, 7),
             panel_grid_major_x=element_blank(),
             ) + 
        scale_fill_manual(values=CBPALETTE)
    )


    p = p + facet_grid('library ~ targeted')
    _ = p.draw()   
    
    plotfile = f'scratch_notebooks/scratch_figs/{stat}_by_site.pdf'
#     os.path.join(config['figs_dir'],
#                         f"targeted_epitope_id_{stat}.pdf")
    print(f"Saving plot to {plotfile}")
    p.save(plotfile)
```

These plots generally look promising. Again, off-target mutation frequency is relatively low, although based on the max mutation frequency plot there may be some individual outliers at off-target sites that are more highly represented. Spread of general targeted mutations looks relatively even. Note that variance across sites could be attributable to some sites having more 'allowed' mutations than others - I haven't dug into that specific analysis here.

The main takeaway here is that the targeted epitope sites are spiked **very** high in mutation frequency, especially for epitopes that were in the 'paired epitopes' pool. We can follow up on this by analyzing how many variants in the library have mutations at these epitope sites:

### Run same analysis on just the targeted epitopes

```python
epitope_mut_counts = mut_counts.loc[mut_counts['epitope'] != 'other']
epitope_mut_counts_stats = (epitope_mut_counts
                            .groupby(['library', 'site'])
                            .aggregate(mut_count_total=('mut_count', 'sum'))
                            .reset_index()
                            .merge(total_variants, how='left', on='library')
                            .drop(['sample'], axis=1)
                            .assign(mut_freq_total=lambda x: x.mut_count_total / x.variant_count)
)

epitope_mut_counts_stats['ha1_site'] = epitope_mut_counts_stats.apply(lambda x: x['site'] - 19, axis=1)
epitope_mut_counts_stats['ha1_site'] = epitope_mut_counts_stats['ha1_site'].astype(str)

epitope_mut_counts_stats
```

Based on this plot, the most highly mutated epitopes are present in around 9% of variants in the library. 

At this point, we've concluded that our mutagenesis strategy was successful for the most part. The rate of off-target mutations in the library is only around 5-7%. Mutation frequency looks pretty even across the gene, and the distribution of variant backgrounds for each individual mutation looks great for our modeling purposes.

### Rate of variants with targeted epitope pairs

```python
# get list of paired epitopes from these primer names
stripped_name = paired_epitope_df['name'].str.findall(r'\d+').tolist()

paired_epitopes = []
for entry in stripped_name:
    entry=list(map(int, entry[1:])) # drop '19' pulled from hk19 in name
    
    # add 3 to account for chimeric numbering conversion
    converted_entry = []
    for site in entry:
        site = site+3
        converted_entry.append(site)
    
    if converted_entry not in paired_epitopes:
        paired_epitopes.append(converted_entry)

paired_epitopes
```

```python
# get full variant dataframe
variant_df = variants.barcode_variant_df
variant_df.head()
```

I think the easiest way to do this is to pull out the site numbers, get a column with the lists of mutated sites, and search for pairs. 

```python
def get_sites(aa_substitutions):
    # pull out every number and return as a list
    sites = list(map(int, re.findall(r'\d+', aa_substitutions)))
    
    return sites

variant_df['mut_sites'] = variant_df.apply(lambda x: get_sites(x['aa_substitutions']), axis=1)
variant_df.head()
```

```python
# Add column that returns 'True' or 'False' depending on whether at least one pair is mutated
def check_pairs(mut_sites, paired_epitopes):
    for pair in paired_epitopes:
        if set(pair).issubset(mut_sites):
            return True
    return False
  
variant_df['epitope_pair'] = variant_df.apply(lambda x: check_pairs(x['mut_sites'], paired_epitopes), axis=1)
```

```python
variant_df.groupby(['library', 'epitope_pair']).size().reset_index().rename(columns={0:'num_variants'})
```

```python
p = (
    ggplot(
        variant_df
        .groupby(['library', 'epitope_pair'])
        .size()
        .reset_index()
        .rename(columns={0:'num_variants'}),
        aes(x='library', y='num_variants', fill='epitope_pair')) + 
    geom_bar(stat='identity', position='dodge') + 
    theme(figure_size=(10, 5),
         panel_grid_major_x=element_blank(),
         ) + 
    scale_fill_manual(values=CBPALETTE))

_ = p.draw() 

plotfile = os.path.join(config['figs_dir'],
                        f"num_variants_w_paired_epi_mut.pdf")
print(f"Saving plot to {plotfile}")
p.save(plotfile)
```

It looks like around 10% of the variants have at least one pair of epitope mutations. So I guess our pooled primer approach works really well, because I did include a ratio of 1:10 targeted : paired_epitope primers. This is much higher than we would ideally want, but it's probably useable. 

Last thing I want to check is the fraction of variants that have each pair of sites mutated:

```python
# get all the variant mutations as list of lists, where each entry is a variant
mut_sites = variant_df['mut_sites'].tolist()

# set up pair counts dict for each pair of epitopes
pair_counts = {}
for entry in paired_epitopes:
    entry = tuple(entry)
    pair_counts[entry] = 0
    
# count instances of variants with each pair mutated
for variant in mut_sites:
    for pair in paired_epitopes:
        if set(pair).issubset(variant):
            pair_counts[tuple(pair)] += 1

# convert counts dict to df for plotting
pair_counts_df = pd.DataFrame.from_dict(pair_counts, orient='index', columns=['count'])
pair_counts_df.index.name = 'pair'
pair_counts_df = pair_counts_df.reset_index()

# convert counts to fraction of total variants
variant_count_sum = total_variants['variant_count'].sum()
pair_counts_df['fraction_variants'] = pair_counts_df.apply(lambda x: x['count']/variant_count_sum, axis=1)

pair_counts_df
```

Note that this is collapsed across **all** libraries, since it made the analysis significantly easier and trends have been very consistent across libraries. 

```python
p = (
    ggplot(
        pair_counts_df,
        aes(x='pair', y='fraction_variants')) + 
    geom_bar(stat='identity', width=0.5) + 
    theme(figure_size=(12, 5),
         panel_grid_major_x=element_blank(),
         ) + 
    scale_fill_manual(values=CBPALETTE))

_ = p.draw() 

plotfile = os.path.join(config['figs_dir'],
                        "frac_variants_w_epitope_pair.pdf")
print(f"Saving plot to {plotfile}")
p.save(plotfile)
```

So it actually looks like my theory about adjacent pairs is incorrect, and for some reason `178/179` is just seen in a really high fraction of variants. 

The only other thing I can think of for the disparity is that I also had targeted primers alongside the random epitope primers, and there were variable numbers of targeted primers depending on how many deleterious mutations were present

```python
# regenerate 'targeted' column, this time *not* blanketing all epitope sites as 'targeted'
# also change labels to 'deleterious' or 'allowed'
# We're basically using this df as a proxy for the full set of targeted mutations bc it's in a format that's
# easy to analyze
mut_counts['targeted'] = np.where((~mut_counts['stripped_mut'].isin(targeted_muts)), 'deleterious', 'allowed')

mut_counts = mut_counts.drop(['stripped_mut'], axis=1)

# cut it down to just the paired epitope sites
mut_counts_paired = mut_counts.loc[mut_counts['epitope'] == 'paired_epitope']
```

```python
mut_counts_paired_targeted = mut_counts_paired.groupby(['library','site', 'targeted']).count().reset_index()
mut_counts_paired_targeted = mut_counts_paired_targeted.loc[mut_counts_paired_targeted['targeted'] == 'allowed'].reset_index()
mut_counts_paired_targeted['site'] = mut_counts_paired_targeted['site'].astype(str)
```

```python
p = (
    ggplot(
        mut_counts_paired
        .loc[mut_counts_paired['library'] == 'libA_rep1'] # should be the same for all libs
        .groupby(['site', 'targeted'])
        .count()
        .reset_index(),
        aes(x='site', y='mutation', fill='targeted')) + 
    geom_bar(stat='identity', position='dodge', width=1) + 
    theme(figure_size=(8, 5),
         panel_grid_major_x=element_blank(),
         ) + 
    scale_fill_manual(values=CBPALETTE[5:]))

_ = p.draw() 
```

The counts for `allowed + deleterious` at each site add up to 20. So we're correctly getting the fractions of allowed vs deleterious mutations.

This isn't the best representation, but we can see that the outliers `178/179`, `211/212`, and `216/217` all have very high ratios of allowed:deleterious mutations. `161` and `163` are the runners up in fraction of paired mutations, and these also have the next highest ratios of allowed:deleterious mutations. Conversely, `209` is very restricted in number of allowed mutations. We see a very low fraction of variants with the `209/211` pair.

Therefore, I think the actual culprit to this skewed representation is the fact that I included both randomized paired epitope primers **and** targeted primers against these same sites. Sites like `178` and `179` had up to 17 targeted primers plus these randomized ones, which is why they're spiked so high.

The one positive takeaway is that we can still expect most of the library to be viable! So this might actually turn out to be a good thing for diminishing returns epistasis analysis, because `178` and `179` are going to be seen in a **lot** of backgrounds and will not just knock out the virus. We can just go ahead and confirm this by plotting the total mutation frequency of allowed vs deleterious mutations at the paired epitope sites:

```python
p = (
    ggplot(
        mut_counts_paired
        .groupby(['library', 'site','targeted'])
        .aggregate(mut_count_total=('mut_count', 'sum'))
        .reset_index()
        .merge(total_variants, how='left', on='library')
        .drop(['sample'], axis=1)
        .assign(mut_freq_total=lambda x: x.mut_count_total/x.variant_count),
        aes(x='site', y='mut_freq_total')) + 
    geom_bar(stat='identity', position='dodge') + 
    theme(figure_size=(10, 5),
         panel_grid_major_x=element_blank(),
         ) + 
    scale_fill_manual(values=CBPALETTE))


p = p + facet_grid('library ~ targeted')
_ = p.draw() 

plotfile = os.path.join(config['figs_dir'],
                        f"rate_deleterious_epitope_muts.pdf")
print(f"Saving plot to {plotfile}")
p.save(plotfile)
```
