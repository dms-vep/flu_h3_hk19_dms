# Analyze the barcode counts
Analyze the barcode counts, and create an aggregated counts files for all samples with adequate counts.

First, import Python modules:

```python
import os

import altair as alt

import pandas as pd

import yaml
```

```python
# allow more rows for Altair
_ = alt.data_transformers.disable_max_rows()
```

Change working directory to top directory of repo:

```python
os.chdir('../')
```

Read configuration:

```python
with open('config.yaml') as f:
    config = yaml.safe_load(f)
```

Read the barcode runs:

```python
barcode_runs = pd.read_csv(config["barcode_runs"])

library_sample_list = barcode_runs["library_sample"].tolist()
print(f"There are {len(library_sample_list)} barcode runs.")
```

## Barcode fates
Get the barcode read "fates" (did they align) for all runs:

```python
fates = (
    pd.concat([
        pd.read_csv(os.path.join(config["barcode_fates_dir"], library_sample + ".csv"))
        for library_sample in library_sample_list
    ])
    .merge(barcode_runs, on=["library", "sample"], validate="many_to_one")
    .drop(columns=["fastq_R1", "notes"])
    .assign(valid=lambda x: x["fate"] == "valid barcode")
)
```

Plot the barcode fates:

```python
sample_types = barcode_runs["sample_type"].unique().tolist()

for date, fate_by_date in fates.groupby('date'):
    fate_by_date_chart = (
        alt.Chart(fate_by_date)
        .encode(
            x=alt.X("count", axis=alt.Axis(format=".2g"), scale=alt.Scale(nice=False)),
            y=alt.Y("fate", title=None),
            color=alt.Color("sample_type", scale=alt.Scale(domain=sample_types)),
            facet=alt.Facet("library_sample", title=None, columns=2),
            fillOpacity="valid",
            tooltip=[
                alt.Tooltip(c, format=".3g") if c == "count" else c
                for c in fate_by_date.columns if c not in ["valid", "library_sample", "sample"]
            ],
        )
        .mark_bar()
        .properties(width=300, height=55, title=f"{date} samples")
    )
    display(fate_by_date_chart)
```

## Read barcode counts
Read counts of all valid and invalid barcodes:

```python
counts = (
    pd.concat([
        pd.read_csv(os.path.join(subdir, library_sample + ".csv")).assign(valid=valid)
        for library_sample in library_sample_list
        for (subdir, valid) in [
            (config["barcode_counts_dir"], True),
            (config["barcode_counts_invalid_dir"], False),
        ]
    ])
)
```

## Average counts per barcode
Next we investigate average barcode counts and the number of barcodes with above-average counts.

Get average invalid and valid barcode counts for each sample, as well as fraction of all barcodes above the average number of counts for valid barcodes:

```python
avg_counts = (
    counts
    .groupby(["library", "sample", "valid"], as_index=False)
    .aggregate(
        mean_counts=pd.NamedAgg("count", "mean"),
        n_barcodes=pd.NamedAgg("barcode", "count"),
    )
    .assign(
        dummy=lambda x: x["valid"] * x["mean_counts"],
        mean_valid_counts=lambda x: x.groupby(["library", "sample"])["dummy"].transform("max"),
    )
    .merge(counts, validate="one_to_many")
    .assign(above_mean_valid=lambda x: x["count"] > x["mean_valid_counts"])
    .groupby(["library", "sample", "valid", "mean_counts", "n_barcodes"], as_index=False)
    .aggregate(n_above_mean_valid=pd.NamedAgg("above_mean_valid", "sum"))
    .assign(
        frac_above_mean_valid=lambda x: x["n_above_mean_valid"] / x["n_barcodes"],
        library_sample=lambda x: x["library"] + "_" + x["sample"],
    )
    .melt(
        id_vars=["library", "sample", "library_sample", "valid"],
        value_vars=["mean_counts", "frac_above_mean_valid"],
    )
    .merge(
        barcode_runs[["library_sample", "date", "virus_batch", "sample_type", "antibody"]],
        on="library_sample",
        validate="many_to_one",
    )
)
```

Now plot these data, showing separate points for barcodes that are "valid" (correspond to known variants) and invalid (do not correspond to known variants).

 - The left panel shows the fraction of all barcodes with more counts than the average for valid barcodes.
 - The right panel shows the average counts for each valid and invalid barcode.
 
The chart is interactive: you can use the drop downs at bottom to select specific subsets, and mouseover points for details.

```python
selections = [
    alt.selection_single(
        fields=[col],
        bind=alt.binding_select(
            options=[None] + avg_counts[col].unique().tolist(),
            labels=["all"] + [str(x) for x in avg_counts[col].unique()],
            name=col,
        )
    )
    for col in ["library", "date", "sample_type", "virus_batch", "antibody"]
]

avg_counts_chart = (
    alt.Chart(avg_counts)
    .encode(
        x="value",
        y=alt.Y("library_sample", title=None),
        color="valid",
        shape="valid",
        column=alt.Column("variable", title=None),
        tooltip=[alt.Tooltip(c, format=".3g") if c == "value" else c
                 for c in avg_counts.columns if c != "library_sample"],
    )
    .mark_point(filled=True, size=50, opacity=0.7)
    .resolve_scale(x="independent")
    .properties(width=150, height=alt.Step(15))
    .configure_axis(labelLimit=500)
    .add_selection(*selections)
)
for selection in selections:
    avg_counts_chart = avg_counts_chart.transform_filter(selection)

avg_counts_chart
```

## Examine invalid barcodes and possible library-to-library contamination
Let's look at the most abundant invalid barcodes.

First, get all invalid barcodes along with where their counts rank among **all** (invalid and valid) barcodes for that library / sample:

```python
ranked_invalid = (
    counts
    .assign(overall_rank=lambda x: (
        x.groupby(["library", "sample"])
        ["count"]
        .transform("rank", ascending=False, method="first")
        .astype(int)
        )
    )
    .query("not valid")
    .merge(barcode_runs)
    [["library", "sample", "barcode", "overall_rank", "count"]]
    .sort_values(["library", "overall_rank"])
)
```

For each library / sample, check if an invalid barcode is in the top 10 most abundant.
If so, print the most abundant invalid barcode:

```python
(
    ranked_invalid
    .query("overall_rank <= 10")
    .groupby(["library", "sample"])
    .aggregate(
        n_invalid_in_top_10=pd.NamedAgg("barcode", "count"),
        top_invalid_rank=pd.NamedAgg("overall_rank", "first"),
        top_invalid_barcode=pd.NamedAgg("barcode", "first"),
    )
)
```

Identify any samples where any of the top 10 barcodes are invalid for that library but valid in another library.
If this happens, it could indicate contamination or sample mixup:

```python
valid_by_library = (
    pd.read_csv(config["virus_variants_w_neut_standard"])
    [["library", "barcode", "target"]]
    .assign(libraries_w_barcode=lambda x: x.groupby("barcode")["library"].transform("nunique"))
)
```

```python
(
    ranked_invalid
    .query("overall_rank <= 10")
    .merge(valid_by_library, on="barcode", suffixes=["", "_other"])
    .groupby(["library", "sample", "library_other"])
    .aggregate(
        n_top_10_from_other_library=pd.NamedAgg("barcode", "count")
    )
)
```

Get fraction of barcode counts that uniquely map to each library for each sample.
If a sample has a lot of barcode counts that map to another library, that is a likely indication of mix-ups.

```python
# get the libraries each barcode belongs to
barcode_by_library = (
    valid_by_library
    .assign(
        barcode_library=lambda x: x["library"].where(
            x["libraries_w_barcode"] == 1,
            x["target"].where(x["target"] != config["gene"],
                x["libraries_w_barcode"].astype(str) + " libraries"
            )
        )
    )
    [["barcode", "barcode_library"]]
    .drop_duplicates()
)
display(
    barcode_by_library
    .groupby("barcode_library")
    .aggregate(n_barcodes=pd.NamedAgg("barcode", "count"))
)
```

```python
# sum counts associated with each library assignment
library_barcode_assignments = (
    counts
    .merge(barcode_by_library, on="barcode", validate="many_to_one")
    .groupby(["library", "sample", "barcode_library"], as_index=False)
    .aggregate(n_counts=pd.NamedAgg("count", "sum"))
    .assign(
        fraction_counts=lambda x: (
            x["n_counts"] / x.groupby(["library", "sample"])["n_counts"].transform("sum")
        )
    )
    .merge(barcode_runs)
    .drop(columns=["fastq_R1", "notes", "antibody_concentration"])
)
```

Here is an interactive plot.
You can click on the legend to control which categories to show, use the dropdown at the bottom to select samples to show, and mouse over bars for details.
You should be worried if an appreciable fraction of any library has counts mapping uniquely to other libraries.
(There could be a bit of legitimate cross-mapping due to barcode sequencing errors and incomplete assignment of barcodes to variants in PacBio, but this should be small).

```python
libs = sorted(library_barcode_assignments["library"].unique())
color_order = (
    libs +
    [c for c in library_barcode_assignments["barcode_library"].unique() if c not in libs]
)

barcode_library_selection = alt.selection_multi(fields=["barcode_library"], bind="legend")

barcode_assignment_chart = (
    alt.Chart(
        library_barcode_assignments
        .assign(order=lambda x: x["barcode_library"].map(lambda s: color_order.index(s)))
    )
    .encode(
        x=alt.X("fraction_counts", scale=alt.Scale(domain=[0, 1])),
        y=alt.Y("library_sample", title=None),
        color=alt.Color("barcode_library", scale=alt.Scale(domain=color_order)),
        order="order",
        tooltip=[
            alt.Tooltip(c, format=".2g") if c in {"n_counts", "fraction_counts"} else c
            for c in library_barcode_assignments.columns
            if c not in {"library_sample"}
        ],
    )
    .mark_bar()
    .properties(width=250, height=alt.Step(15))
    .configure_axis(labelLimit=500)
    .add_selection(barcode_library_selection, *selections)
    .transform_filter(barcode_library_selection)
)
for selection in selections:
    barcode_assignment_chart = barcode_assignment_chart.transform_filter(selection)


barcode_assignment_chart
```

## Fraction of counts from neutralization standard barcodes
Get and plot the fraction of all valid counts from neutralization standard barcodes:

```python
neut_standard_barcodes = pd.read_csv(config["neut_standard_barcodes"])["barcode"].tolist()

print(f"Read {len(neut_standard_barcodes)} neutralization-standard barcodes")
```

```python
frac_neut_standard = (
    counts
    .query("valid")
    .assign(is_neut_standard=lambda x: x["barcode"].isin(neut_standard_barcodes))
    .groupby(["library", "sample", "is_neut_standard"], as_index=False)
    .aggregate(counts=pd.NamedAgg("count", "sum"))
    .assign(
        total_counts=lambda x: x.groupby(["library", "sample"])["counts"].transform("sum"),
        frac=lambda x: x["counts"] / x["total_counts"],
    )
    .query("is_neut_standard")
    .merge(barcode_runs, validate="one_to_one")
    .drop(columns=["fastq_R1", "notes", "antibody_concentration", "counts", "is_neut_standard"])
)
```

## Write counts to files for samples with sufficient average counts

Get the valid barcode counts for samples with adequate counts:

```python
valid_counts = counts.query("valid").drop(columns="valid")

avg_valid_counts = (
    valid_counts
    .groupby(["library", "sample"], as_index=False)
    .aggregate(avg_counts=pd.NamedAgg("count", "mean"))
)
print(f"Examining average valid counts for all {len(avg_valid_counts)} samples")

min_avg_counts = config["min_avg_counts"]
print(f"Samples with <{min_avg_counts} average counts per variant are being discarded")
display(
    avg_valid_counts
    .query("avg_counts < @min_avg_counts")
    .sort_values("avg_counts")
    .reset_index(drop=True)
)

valid_counts = (
    valid_counts
    .merge(avg_valid_counts.query("avg_counts >= @min_avg_counts"), how="right")
    .drop(columns=["avg_counts"])
)
n_retained = len(valid_counts[["library", "sample"]].drop_duplicates())
print(f"Retaining {n_retained} samples with adequate counts")
```

Write the counts for the retained samples to a file:

```python
print(f"Writing counts for valid barcodes to {config['barcode_counts']}")

valid_counts.to_csv(config["barcode_counts"], index=False)
```
