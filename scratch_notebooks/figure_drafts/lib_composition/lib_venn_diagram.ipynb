```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt
```

```python
# Define the counts
deleterious_mutations_count = 7077
library_mutations_count = 2505
overlap_count = 44
epitope_mutations_count = 418

# Create a Venn diagram instance
venn = venn3(subsets=(deleterious_mutations_count, library_mutations_count, overlap_count, epitope_mutations_count, 0, 0, 0),
             set_labels=('Deleterious Mutations', 'Library Mutations', 'Epitope Mutations'))

# Set the circle sizes based on counts
venn.get_patch_by_id('100').set_alpha(1.0)
venn.get_patch_by_id('010').set_alpha(1.0)
venn.get_patch_by_id('001').set_alpha(1.0)

# Customize the labels
venn.get_label_by_id('100').set_text(deleterious_mutations_count)
venn.get_label_by_id('010').set_text(library_mutations_count)
venn.get_label_by_id('001').set_text(epitope_mutations_count)
venn.get_label_by_id('110').set_text(overlap_count)

# Show the Venn diagram
plt.title("Venn Diagram with Proportional Circle Sizes")
plt.savefig("venn_diagram.pdf", format="pdf", dpi=300)
```
